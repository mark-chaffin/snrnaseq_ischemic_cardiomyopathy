# Step 4: Additional sub-clustering QC of global map
Even after removing low-quality clusters and nuclei, we are often left with some misclassified/low quality nuclei in each cluster. Typically, a sub-clustering procedure can help to pull out nuclei that are enriched in MT reads, or highly express markers from other clusters (likely missed doublets). We will do an iterative sub-clustering procedure at varying resolutions and then look for misclassified sub-clusters by scoring nuclei based on the profiles of the global clusters.

### Find marker genes of the global cell types
We will find misclassified/bad nuclei in each of our global clusters upon sub-clustering. It will be good to filter those out here. To identify misclassified/bad nuclei, we can score all cells for "profiles" of the global clusters and look for subclusters that are enriched in scores from other, unrelated cell types.

```python
# Read in map with leiden clustering and RAW counts
adata = sc.read(os.path.join(path_aggregation,'humanICM.PostQC.Map/PostQC.Neighbor.UMAP.h5ad'))
```

```python
# Now normalize prior to PCA (and take log)
# We use a normalization factor of 1e4 as is done in Seurat and scanpy default tutorials
# but there are other reasonable choices here
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
# Group into major cell types -- this is collapsing some information
# We are collapsing a few clusters here that are very related
cluster_map = {'0':'cardiomyocyte', '1':'fibroblast', '2':'endothelial', '3':'pericyte',
               '4':'macrophage', '5':'endothelial', '6':'lymphocyte', '7':'vsmc',
               '8':'lymphatic_endothelial','9':'adipocyte','10':'neuronal',
               '11':'endothelial','12':'prof_macrophage','13':'mast_cell',
               '14':'cardiomyocyte','15':'epicardial'}
adata.obs['cluster'] = [cluster_map[x] for x in adata.obs['leiden0.5']]
```

```python
# Do a basic wilcoxon test comparing expression of each gene in the target cluster
# to expression in all other clusters. We can derive an AUC measure from this.
sc.tl.rank_genes_groups(adata, 'cluster', method='wilcoxon', n_genes = adata.shape[1])
```

```python
# Build a dataframe of the wilcoxon test
allmerge = []
for q in range(len(adata.obs['cluster'].unique())):
    forframe = {}
    for i in adata.uns['rank_genes_groups'].keys():
        if i =='params':
            continue
        newcol = []
        sub = adata.uns['rank_genes_groups'][i].tolist()
        for s in sub:
            newcol.append(s[q])
        forframe[i] = newcol
    wilcox = pd.DataFrame(forframe)
    wilcox['cluster'] = adata.uns['rank_genes_groups']['names'].dtype.names[q]
    allmerge.append(wilcox)
fullresults = pd.concat(allmerge)
```

```python
# Calculate the number of nuclei in the target cluster and all other clusters
# We will use this to derive an AUC from the wilcoxon test statistic
nlookup = pd.DataFrame(adata.obs['cluster'].value_counts()).reset_index()
nlookup.columns = ['cluster', 'n1']
nlookup['n2'] = adata.shape[0] - nlookup['n1']
```

```python
# Merge in the cluster counts
fullresults = fullresults.merge(nlookup, on = 'cluster', how='left')

# Calculate the U statistic from the wilcoxon ranked sum test W statistic
fullresults['U'] = (fullresults['scores'] * np.sqrt((fullresults['n1']*fullresults['n2']*(fullresults['n1'] + fullresults['n2'] + 1)) / 12)) + ((fullresults['n1'] * fullresults['n2']) / 2)

# Convert the U-statistic into an AUC
fullresults['AUC'] = fullresults['U']/(fullresults['n1'] * fullresults['n2'])
```

```python
# Save the marker gene lists
os.system('mkdir ' + os.path.join(path_aggregation,'humanICM.PostQC.Map/marker_genes_for_outliers'))
fullresults.to_csv(os.path.join(path_aggregation,'humanICM.PostQC.Map/marker_genes_for_outliers/marker_genes.txt'),
                   sep='\t',index=None,na_rep='NA')
```

### Sub-cluster each of the major cell types by increasing resolution

```python
# Read in map with leiden clustering and RAW counts
adata = sc.read(os.path.join(path_aggregation,'humanICM.PostQC.Map/PostQC.Neighbor.UMAP.h5ad'))
```

```python
# Now normalize prior to PCA (and take log)
# We use a normalization factor of 1e4 as is done in Seurat and scanpy default tutorials
# but there are other reasonable choices here
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

**Sub-clustering procedure:**
1. Subset to just cells from a given cell type
2. Calculate a new neighborhood graph
3. Iterate over different ranges of leiden resolution (increments of 0.05) until at least one cluster emerges that doesn't have a strong marker (AUC > 0.6) compares to the other clusters. This typically happens when we start pulling out uninformative clusters.
4. Save the new sub-cluster results.

```python
# Define a sub-cluster procedure
def subcluster(adata, cluster, cluster_key, auc_stop, dim_key,
               n_dim, n_neighbor, incr, min_res, max_res):
    '''
    adata: adata object
    cluster: cluster to sub-cluster
    cluster_key: the variable in .obs storing the clusters
    auc_stop: a cut-off point for AUC to consider a gene a marker
    dim_key: the key for the PCs to construct a neighborhood graph from
    n_dim: the number of dimensions to use in sub-clustering
    n_neighbor: the number of neighbors to use in sub-clustering
    incr: the increment size for the clustering resolution
    min_res: the starting (minimum) resolution
    max_res: the maximum resolution to sub-cluster to
    '''
    incr = round(incr, 2)
    min_res = round(min_res,2)
    max_res = round(max_res,2)
    
    # Subset to just the cluster of interest and recalculate neighbors
    bdata = adata[adata.obs[cluster_key].isin(cluster)].copy()
    print('Clustering ' + str(bdata.shape[0]) + ' nuclei')
    print('Calculating new neighborhood graph...')
    sc.pp.neighbors(bdata, n_neighbors=n_neighbor, n_pcs=n_dim, use_rep=dim_key)
    
    # iterate through various resolutions
    marker_res = {}
    for i in range(int(min_res*100),int(max_res*100),int(incr*100)):
        res = i/100.
        print(res)
        sc.tl.leiden(bdata, resolution = res, key_added = 'leiden'+str(res))
        print(len(bdata.obs['leiden'+str(res)].unique()))
        if len(bdata.obs['leiden'+str(res)].unique()) <= 1:
            prev = res
            continue
        # find markers
        sc.tl.rank_genes_groups(bdata, 'leiden' + str(res), method='wilcoxon', n_genes = bdata.shape[0])
        # Convert markers to datatable
        allmerge = []
        for q in range(len(bdata.obs['leiden' + str(res)].unique())):
            forframe = {}
            for i in bdata.uns['rank_genes_groups'].keys():
                if i =='params':
                    continue
                newcol = []
                sub = bdata.uns['rank_genes_groups'][i].tolist()
                for s in sub:
                    newcol.append(s[q])
                forframe[i] = newcol
            wilcox = pd.DataFrame(forframe)
            wilcox['cluster'] = bdata.uns['rank_genes_groups']['names'].dtype.names[q]
            allmerge.append(wilcox)
        fullresults = pd.concat(allmerge)
        nlookup = pd.DataFrame(bdata.obs['leiden' + str(res)].value_counts()).reset_index()
        nlookup.columns = ['cluster', 'n1']
        n2_v = []
        for i in nlookup['cluster']:
            n2_v.append(nlookup[nlookup['cluster']!=i]['n1'].sum())
        nlookup['n2'] = n2_v
        nlookup['cluster'] = [x for x in nlookup['cluster']]
        fullresults = fullresults.merge(nlookup, on = 'cluster', how='left')
        fullresults['U'] = (fullresults['scores'] * np.sqrt((fullresults['n1']*fullresults['n2']*(fullresults['n1'] + fullresults['n2'] + 1)) / 12)) + ((fullresults['n1'] * fullresults['n2']) / 2)
        fullresults['AUC'] = fullresults['U']/(fullresults['n1'] * fullresults['n2'])
        marker_res[res] = fullresults
        # stopping criteria
        if len(fullresults[fullresults['AUC'] > auc_stop]['cluster'].value_counts().index) < len(bdata.obs['leiden'+str(res)].unique()):
            break
        prev = res
    if not prev in marker_res.keys():
        return(bdata.obs[['leiden'+str(prev)]], None)
    else:
        return(bdata.obs[['leiden'+str(prev)]], marker_res[prev])
```

```python
os.system('mkdir ' + os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering'))
```

```python
# Set some universal parameters we'll use for every cell type
cluster_key = 'leiden0.5'
auc_stop = 0.6
dim_key = 'PC_harmony'
n_dim = 50
n_neighbor = 15
incr = .05
min_res = 0.05
max_res = 1.0

# Loop through each cell type and sub-cluster
for name,cluster in zip(['cardiomyocyte', 'fibroblast', 'endothelial',
                         'pericyte', 'macrophage', 'lymphocyte', 'vsmc',
                         'lymphatic_endothelial', 'adipocyte', 'neuronal',
                         'prof_macrophage', 'mast_cell', 'epicardial'],
                        [['0','14'], ['1'], ['2','5','11'], ['3'], ['4'], ['6'],
                         ['7'], ['8'], ['9'], ['10'], ['12'], ['13'], ['15']]):
    print(name)
    # Run sub-clustering
    labels, markers = subcluster(adata, cluster, cluster_key, auc_stop, dim_key, n_dim, n_neighbor, incr, min_res, max_res)
    labels.reset_index().to_csv(os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering/'+name+'_labels.txt'),
                                sep='\t', index=None)

    # If there are some markers, save them
    if not markers is None:
        markers.to_csv(os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering/'+name+'_markers.txt'),
                       sep='\t', index=None)

    # Look at sub-clustering
    print(labels.columns[0])
    print(labels[labels.columns[0]].value_counts())
    lab_map = {x:y for x,y in zip(labels.index, labels[labels.columns[0]])}
    pdata = adata[adata.obs[cluster_key].isin(cluster)]
    pdata.obs['Sub_cluster'] = [lab_map[x] for x in pdata.obs.index]
    glist = [i for j in [markers[markers['cluster']==x].sort_values('AUC',ascending=False).head(5)['names'].tolist() for x in markers['cluster'].unique()]
             for i in j]
    sc.pl.dotplot(pdata, glist, groupby='Sub_cluster', standard_scale='var')
    sc.pl.umap(pdata, color = 'Sub_cluster', palette=sc.pl.palettes.vega_20)
```

### Scoring sub-clusters based on expression of non-target cluster marker genes
We will identify populations enriched for other cell type markers and flag them in the H5AD -- these are often low quality or misclassified cells.

```python
# Read in map with leiden clustering and RAW counts
adata = sc.read(os.path.join(path_aggregation,'humanICM.PostQC.Map/PostQC.Neighbor.UMAP.h5ad'))
```

```python
def contam_misclass_subclust(adata, cluster_name, cluster, cluster_key,
                             marker_file, subclusters, marker_cluster_key, n_markers,
                             output_dir):
    '''
    adata: the adata object
    cluster_name: the cell type name
    cluster: a list of the qualifying groups of interest
    cluster_key: the cluster key in the .obs corresponding the cluster identities
    marker_file: the file with the global marker genes
    subclusters: the file containing the sub-cluster labels for the cell type
    marker_cluster_key: the key matching the cluster_name in the above
    n_markers: the number of markers to use in scoring
    output_dir: the output directory
    '''
    pdata = adata[adata.obs[cluster_key].isin(cluster)].copy()
    # Normalize and scale -- required before scoring
    print('Normalizing...')
    sc.pp.normalize_total(pdata, 1e4)
    sc.pp.log1p(pdata)
    sc.pp.scale(pdata)
    
    # labels -- derived in the above subclustering procedure for each cell type
    print('Labeling...')
    subcluster_labels = pd.read_csv(os.path.join(subclusters,cluster_name+'_labels.txt'), sep='\t')
    label_dict = {x:y for x,y in zip(subcluster_labels['index'], subcluster_labels[subcluster_labels.columns[1]])}
    pdata.obs['subcluster'] = [label_dict[x] for x in pdata.obs.index]
    
    # markers -- derived these for the global cell types above
    print('Getting markers...')
    markers = pd.read_csv(marker_file, sep='\t')
                            
    # For each cell type, score all cells in the cluster based on the top n_markers of that cluster based on AUC
    print('Scoring...')
    for i in markers[marker_cluster_key].unique():
        submark = markers[markers[marker_cluster_key]==i].sort_values('AUC', ascending=False).head(n_markers)
        sc.tl.score_genes(pdata, submark['names'].tolist(),
                          ctrl_size = n_markers, score_name = 'score_'+str(i))
    # An additional scoring based on MT genes
    sc.tl.score_genes(pdata, [x for x in pdata.var.index if re.search('MT-',x)],
                      ctrl_size = n_markers, score_name = 'score_mt') 
    # Save a file with the scoring of nuclei from the global profiles
    out = pdata.obs[['subcluster'] + ['score_' + str(y) for y in markers[marker_cluster_key].unique().tolist() + ['mt']]]
    out.to_csv(os.path.join(output_dir,cluster_name+'_scored.txt.gz'), sep='\t', compression='gzip')
    # Plot the distribution of the cell type scoring across the subclusters using box plots
    nrow = math.ceil((len(markers[marker_cluster_key].unique()) + 1)/3.)
    f,ax = plt.subplots(nrow, 3, figsize=(14,nrow*2.5))
    f.subplots_adjust(hspace= .5)
    x = -1
    y = 0
    for i in markers[marker_cluster_key].unique().tolist() + ['mt']:
        x += 1
        if x == 3:
            y += 1
            x = 0
        sns.boxplot(x = 'subcluster', y = 'score_'+str(i), data = pdata.obs, ax = ax[y][x])
        ax[y][x].set_title('score_'+str(i))
    plt.show()
    return(out)
```

```python
# Set the cell types and the leiden0.5 groups that wend into each cluster
celltypes = {'cardiomyocyte':['0','14'], 'fibroblast':['1'], 'endothelial':['2','5','11'],
             'pericyte':['3'], 'macrophage':['4'], 'lymphocyte':['6'], 'vsmc':['7'],
             'lymphatic_endothelial':['8'], 'adipocyte':['9'], 'neuronal':['10'],
             'prof_macrophage':['12'],'mast_cell':['13'], 'epicardial':['15']}
```

```python
# Make output directory for the scoring
os.system('mkdir ' + os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering/subcluster_scoring'))
```

```python
# Score the nuclei for each subcluster based on all the global profiles
# Look at the distribution of these score across sub-clusters to identify
# misclassified subclusters
for ct in reversed([y for y in celltypes.keys()]):
    print(ct)
    res = contam_misclass_subclust(adata,
                             ct,
                             celltypes[ct],
                             'leiden0.5',
                             os.path.join(path_aggregation,'humanICM.PostQC.Map/marker_genes_for_outliers/marker_genes.txt'),
                             os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering'),
                             'cluster',
                             50,
                             os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering/subcluster_scoring'))
```

```python
# Based on inspection of these scoring plots above, the following "low-quality" sub-clusters
# were flagged for each global cluster
score_path = os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering')
incorp = {'cardiomyocyte': ('CM',[]), # No real misclass here
          'fibroblast':('FB',[]), # No misclass here as only 1 'subcluster'
          'endothelial':('EC',[]), # No real misclass here
          'pericyte':('PC',[1]), # 1 looks like FB
          'macrophage':('MP',[]), # No real misclass here
          'vsmc':('VSMC',[3,4]), # 3 looks like FB, 4 looks like MP
          'lymphocyte':('LC',[1,2,3]), # 1 looks like EC, 2 looks like FB/EpiC, 3 looks like PC/VSMC
          'adipocyte':('AD',[1,2,3,4]), # 1 looks like FB, 2 looks like CM, 3 looks like MP, 4 looks like EpiC
          'neuronal':('NRN',[1,2,3]), # 1 looks like FB, 2 looks like MAC, 3 looks like EC
          'lymphatic_endothelial':('LEC',[1,2]), # 1 looks like FB, 2 looks like MT
          'mast_cell':('MAST',[0,3,4,5]), # 0 looks like CM, 3 looks like EC, 4 looks like FB, and 5 looks like MAC
          'prof_macrophage':('ProfMP',[1,2,3]), # 1 looks like EC, 2 looks like fibroblasts, 3 looks pericyte/vsmc
          'epicardial':('EpiC',[0,3])} # 0 looks like CMs and 3 has high MT
```

```python
# Create a key to append the sub-cluster and flag to the adata object
# See how many sub-clusters are removed from each
subcluster_dict = {}
ct = 0
ct2 = 0
for celltype in incorp.keys():
    label = pd.read_csv(os.path.join(path_aggregation,'humanICM.PostQC.Map/SubClustering/'+celltype+'_labels.txt'), sep='\t')
    label['flag'] = label[label.columns[1]].isin(incorp[celltype][1])
    label['flag'] = [1 if x else 0 for x in label['flag']]
    ct += label['flag'].sum()
    ct2 += label.shape[0]
    label['subcluster'] = [incorp[celltype][0]+'-'+str(x) for x in label[label.columns[1]]]
    print(celltype, str(label[label['flag']==1].shape[0]) + '/' +  str(label.shape[0]) + ' (' + \
                    str(round(label[label['flag']==1].shape[0] / label.shape[0], 3) * 100) + '%)')
    for barcode, subcluster, flag in zip(label['index'], label['subcluster'], label['flag']):
        subcluster_dict[barcode] = (subcluster,flag)
```

```python
# OVerall this is removing a very small number of overall nuclei
print(str(ct) + '/' + str(ct2) + '(' + str(round(ct/ct2 * 100,3)) + '%)')
```

```python
# Add sub-cluster and sub-cluster flag to adata
adata.obs['subcluster'] = [subcluster_dict[x][0] for x in adata.obs.index]
adata.obs['subcluster_flag'] = [subcluster_dict[x][1] for x in adata.obs.index]
```
