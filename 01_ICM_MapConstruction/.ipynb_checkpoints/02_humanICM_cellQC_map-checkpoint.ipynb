# Step 2: Aggregate all non-empty droplets and remove low quality nuclei
The goal here is to aggregate all non-empty droplets across all experiments, accounting for inter-sample heterogeneity by aligning using Harmony. We will then look for low quality clusters within this alignment based on QC metrics including:
1. Percent MT reads
2. Proportion of reads mapping to exonic regions
3. Doublet score
4. Entropy

After removing low quality clusters, we can perform a nuclei QC procedure on a per sample, per cell type basis to account for heterogeneity between cell types on QC metrics, and heterogeneity between samples.

### Nuclei QC Procedure for Remaining Clusters
Beyond removing these entire clusters of low quality nuclei described above, we also perform QC by cluster/sample combination to ensure that we only retain high quality nuclei. We perform QC by cluster/sample because some cell types are intrinsically more transcriptionally complex (e.g. Cardiomyocytes) so building QC distributional across all cells simultaneously disproportionately will remove the less complex cell types (e.g. Lymphocytes). Likewise, we will see a similar issue because of differing sequencing depth in each sample. Building distributional cutoffs for QC metrics by cluster/sample helps minimize these concerns.

**The per-sample/cluster QC procedure is laid out in the follow functions:**

This could be optimized to be more efficient and more cleanly written, but is functional as is.

```python
# Fucntion to determine IQR outliers based on:
# < Q3 - mult*IQR
# > Q1 + mult*IQR
# Specify whether whether to look for high, low, or both
def iqr_out(vector, mult, high, low):
    '''
    vector: QC metric vector
    mult: What is the outlier multiplier for the IQR (generally 1.5)
    high: Look for outliers on the high side (True/False)
    low: Look for outliers on the low side (True/False)
    '''
    q1 = np.percentile(vector, 25)
    q3 = np.percentile(vector, 75)
    iqr = q3 - q1
    if high and low:
        outliers = (vector > (q3 + mult * iqr)) | (vector < (q1 - mult * iqr))
    if high and not low:
        outliers = (vector > (q3 + mult * iqr))
    if not high and low:
        outliers = (vector < (q1 - mult * iqr))
    if not high and not low:
        outliers = np.nan
    return(outliers)
```

```python
# Function to perform QC a selection of cells -- based on the idea that we will do per sample/cluster combination
def per_sample_cluster_qc(adata, cluster, sample, cluster_key, sample_key):
    '''
    adata: AnnData object
    cluster: Specific cluster ID
    sample: Specific sample ID
    cluster_key: The key in the adata.obs for cluster
    sample_key: The key in adata.obs for sample
    '''
    # Subset to the specific sample/cluster combination
    sub = adata[(adata.obs[cluster_key]==cluster)&(adata.obs[sample_key]==sample)].copy()
    # If the particular sample/cluster pair has > 30 cells, QC based on distribution
    # If <= 30 cells, do QC based on hard thresholds as distributions are less reliable
    # Strict cutoff of %MT <=0.05, n_genes >= 150, nUMI >= 150 regardless
    if sub.shape[0] > 30:
        sub.obs['cellbender_ncount_outlier'] = iqr_out(sub.obs['cellbender_ncount'], 1.5, True, True) # both high and low nUMI
        sub.obs['cellbender_ngenes_outlier'] = iqr_out(sub.obs['cellbender_ngenes'], 1.5, True, True) # both high and low nGenes
        sub.obs['cellranger_percent_mito_outlier'] = iqr_out(sub.obs['cellranger_percent_mito'], 1.5, True, False) # Only high %MT
        sub.obs['cellbender_entropy_outlier'] = iqr_out(sub.obs['cellbender_entropy'], 1.5, True, True) # Both high and low entropy
        sub.obs['exon_prop_outlier'] = iqr_out(sub.obs['exon_prop'], 1.5, True, False) # Only high proportion exonic
        sub.obs['cellranger_doublet_scores_outlier'] = iqr_out(sub.obs['cellranger_doublet_scores'], 1.5, True, False) # Only high doublet score
        sub.obs['log_gene_entropy_outlier'] = iqr_out(sub.obs['log_gene_entropy'], 1.5, True, True) # Both high/low log(gene*entropy)
        sub.obs['outlier'] = sub.obs['cellbender_ncount_outlier'] | \
                             sub.obs['cellbender_ngenes_outlier'] | \
                             sub.obs['cellranger_percent_mito_outlier'] | \
                             sub.obs['cellbender_entropy_outlier'] | \
                             sub.obs['exon_prop_outlier'] | \
                             sub.obs['cellranger_doublet_scores_outlier'] | \
                             sub.obs['log_gene_entropy_outlier'] | \
                             (sub.obs['cellbender_ngenes'] < 150) | \
                             (sub.obs['cellbender_ncount'] < 150) | \
                             (sub.obs['cellranger_percent_mito'] > 0.05)
    # The hard cutoffs when not many nuclei
    else:
        sub.obs['cellbender_ncount_outlier'] = sub.obs['cellbender_ncount'] > 15000 # Only high here because low is covered later
        sub.obs['cellbender_ngenes_outlier'] = sub.obs['cellbender_ngenes'] > 6000 # Only high here because low is covered later
        sub.obs['cellranger_percent_mito_outlier'] = sub.obs['cellranger_percent_mito'] > 0.05
        sub.obs['cellbender_entropy_outlier'] = sub.obs['cellbender_entropy'] < 8.
        sub.obs['exon_prop_outlier'] = sub.obs['exon_prop'] > 0.18
        sub.obs['cellranger_doublet_scores_outlier'] = sub.obs['cellranger_doublet_scores'] > 0.30
        sub.obs['log_gene_entropy_outlier'] = sub.obs['log_gene_entropy'] > 75
        sub.obs['outlier'] = sub.obs['cellbender_ncount_outlier'] | \
                             sub.obs['cellbender_ngenes_outlier'] | \
                             sub.obs['cellranger_percent_mito_outlier'] | \
                             sub.obs['cellbender_entropy_outlier'] | \
                             sub.obs['exon_prop_outlier'] | \
                             sub.obs['cellranger_doublet_scores_outlier'] | \
                             sub.obs['log_gene_entropy_outlier'] | \
                             (sub.obs['cellbender_ngenes'] < 150) | \
                             (sub.obs['cellbender_ncount'] < 150) | \
                             (sub.obs['cellranger_percent_mito'] > 0.05)
    return(sub.obs)
```

```python
# Function to QC a sample
def qc_sample(adata, sample, cluster_key, sample_key, bad_clusters):
    '''
    adata: AnnData object
    sample: Specific sample ID
    cluster_key: The key in the adata.obs for cluster
    sample_key: The key in adata.obs for sample
    bad_clusters: A list of the low-quality clusters determined previously
    '''
    tocat = []
    # Loop through each cluster in the sample and determine which nuclei to remove
    for cl in adata[adata.obs[sample_key]==sample].obs[cluster_key].unique():
        res = per_sample_cluster_qc(adata, cl, sample, cluster_key, sample_key)
        tocat.append(res)
    result1 = pd.concat(tocat)
    # Set outliers based on the above test
    # or if the cluster was deemed a "bad cluster" in the global analysis
    result1['outlier'] = result1['outlier'] | result1[cluster_key].isin(bad_clusters)
    # Calculate DE genes between outliers and retained by clusters (basic t-test for speed)
    # Exclude the "bad cluster" for this since we're getting rid of ALL of these nuclei
    # This is a sanity check to ensure we aren't removing biologically relevant nuclei!
    # Mostly should be MT genes and other clear contaminants that differ between groups
    sub = adata[(adata.obs[sample_key]==sample)&(~adata.obs[cluster_key].isin(bad_clusters))].copy()
    sub.obs['outlier'] = sub.obs.index.isin(result1[result1['outlier']].index)
    sc.pp.normalize_total(sub, target_sum=1e4)
    sc.pp.log1p(sub)
    sub.obs['outlier_cat'] = ['outlier' if x else 'non-outlier' for x in sub.obs['outlier']]
    sub.obs['outlier_cat'] = sub.obs['outlier_cat'].astype('category')
    cellcounts = pd.crosstab(sub.obs[cluster_key],sub.obs['outlier'])
    cellcounts = cellcounts[(cellcounts[False] > 10)&(cellcounts[True] > 10)] #only t-test if we have >10 nuclei in each group
    # Do a rough t-test to compare retained nuclei to removed nuclei
    # Loop through each cluster individuall here
    sc.tl.rank_genes_groups(sub, cluster_key, method='t-test', key_added='cluster_markers', n_genes=10000,
                            groups = cellcounts.index.tolist())
    tocat = []
    for cl in cellcounts.index:
        sub2 = sub[sub.obs[cluster_key]==cl].copy()
        sc.tl.rank_genes_groups(sub2, 'outlier_cat', method='t-test', key_added='s1', n_genes=10000)
        forframe = {}
        for i in sub2.uns['s1'].keys():
            if i =='params':
                continue
            newcol = []
            subset = sub2.uns['s1'][i].tolist()
            for s in subset:
                newcol.append(s[1])
            forframe[i] = newcol
        ttest = pd.DataFrame(forframe)
        ttest['cluster'] = cl
        tocat.append(ttest)
    ttest = pd.concat(tocat)
    return(result1,ttest,sub)
```

```python
def plot_sample(res, sample, cluster_key, save_fig):
    '''
    res: The output from qc_sample
    sample: Specific sample ID
    sample_key: The key in adata.obs for sample
    save_fig: Root directory to specify where to save the QC processing JPG
    '''
    f = plt.figure(num=None, figsize=(22,8), dpi=100, facecolor='w', edgecolor='k')
    gs = gridspec.GridSpec(8,14)
    gs.update(left=0.1, right=1, wspace=2.0, hspace=1.5)
    ax1 = plt.subplot(gs[0:2,0:4])
    ax2 = plt.subplot(gs[0:2,4:8])
    ax3 = plt.subplot(gs[2:4,0:4])
    ax4 = plt.subplot(gs[2:4,4:8])
    ax5 = plt.subplot(gs[4:6,0:4])
    ax6 = plt.subplot(gs[4:6,4:8])
    ax7 = plt.subplot(gs[6:8,0:4])
    ax8 = plt.subplot(gs[0:4,8:11])
    ax9 = plt.subplot(gs[0:4,11:14])
    ax10 = plt.subplot(gs[4:8,8:14])
    # Boxplots of QC metrics, by cell type
    # Coloring the deemed outlier nuclei in each cluster
    res[cluster_key] = [int(x) for x in res[cluster_key]]
    for x,i in zip([ax1,ax2,ax3,ax4,ax5,ax6,ax7],
                     ['cellbender_ncount','cellbender_ngenes','cellranger_percent_mito','cellbender_entropy',
                      'exon_prop','cellranger_doublet_scores','log_gene_entropy']):
        sns.boxplot(x = res[cluster_key], y = res[i],  ax = x, fliersize=0)
        x.scatter(res[cluster_key].astype('category').cat.codes[res[i+'_outlier']],
                  res[res[i+'_outlier']][i], color = 'red',
                  s = 5)
    # UMAP showing the nuclei removal
    ax8.scatter(res['UMAP1'], res['UMAP2'], s = 1, edgecolor=None,
                color = ['red' if x else 'grey' for x in res['outlier']])
    ax8.set_xlabel('UMAP1')
    ax8.set_ylabel('UMAP2')
    ax8.set_title('Cell Removal')
    # Counts of cells removed
    ax9.set_xlim(0,1)
    ax9.set_ylim(0,1)
    ax9.spines['top'].set_visible(False)
    ax9.spines['right'].set_visible(False)
    ax9.spines['left'].set_visible(False)
    ax9.spines['bottom'].set_visible(False)
    ax9.set_xticks([])
    ax9.set_yticks([])
    ax9.text(0,.5,'Cells Removed = ' + str(res[res['outlier']].shape[0]), fontsize = 20)
    ax9.text(0,.4,'Cells Retained = ' + str(res[~res['outlier']].shape[0]), fontsize = 20)
    # Barplot showing removal by cluster -- just to make sure it seems fair
    ax10.bar(x = [int(x) for x in res[['outlier',cluster_key]].groupby(cluster_key).mean().index.tolist()],
            height = res[['outlier',cluster_key]].groupby(cluster_key).mean()['outlier'].tolist())
    ax10.set_xticks([int(x) for x in res[['outlier',cluster_key]].groupby(cluster_key).mean().index.tolist()])
    for i,j in zip(res[cluster_key].value_counts(),
                   res[cluster_key].value_counts().index):
        plt.text(int(j)-.2,
                 1.1,
                 str(res[(res[cluster_key]==j)&(res['outlier'])].shape[0]) + '/' + str(i),
                 rotation = 90)
    ax10.set_ylim(0,1.5)
    ax10.set_xlabel('Cluster')
    ax10.set_ylabel('Fraction of Cluster removed')
    # Title
    plt.suptitle("SAMPLE: " + sample, fontsize=26)
    f.savefig(save_fig, bbox_inches='tight')
    plt.show()
```

```python
# The complete pipeline linking the above functions together
def qc_pipeline(adata, cluster_key, sample_key, bad_clusters, savefig):
    '''
    adata: AnnData object
    cluster_key: The key in the adata.obs for cluster
    sample_key: The key in adata.obs for sample
    bad_clusters: A list of the low-quality clusters determined previously
    savefig: Root directory to specify where to save the QC processing JPG
    '''
    tocat = []
    for sample in adata.obs[sample_key].unique():
        print('Processing sample ' + str(sample))
        outcells,de_test,mysub = qc_sample(adata, sample, cluster_key, sample_key, bad_clusters)
        mysub.obs['plot_group'] = [x + '_' + y for x,y in zip(mysub.obs[cluster_key],mysub.obs['outlier_cat'])]
        mysub.obs['plot_group'] = mysub.obs['plot_group'].astype('category')
        print('Outlier detection\n')
        print(outcells['outlier'].value_counts())
        print('Looking at DE between outliers and remaining by cluster...')
        tcat = []
        for en,cl in enumerate(mysub.uns['cluster_markers']['names'].dtype.names):
            forframe = {}
            for i in mysub.uns['cluster_markers'].keys():
                if i =='params':
                    continue
                newcol = []
                subset = mysub.uns['cluster_markers'][i].tolist()
                for s in subset:
                    newcol.append(s[en])
                forframe[i] = newcol
            ttest = pd.DataFrame(forframe)
            ttest['cluster'] = cl
            ttest = ttest[ttest['pvals_adj'] < 0.05]
            tcat.append(ttest)
        ttest = pd.concat(tcat)
        for i in de_test['cluster'].unique():
            toplot = de_test[(de_test['pvals_adj'] < 0.05)&\
                             (de_test['cluster']==i)&\
                             (de_test['names'].isin(ttest[ttest['cluster']==i]['names']))&\
                             (de_test['logfoldchanges'] > 0.5)].head(20)['names'].tolist()
            if len(toplot) == 0:
                continue
            sc.pl.dotplot(mysub[mysub.obs[cluster_key] == i],
                          toplot, groupby='plot_group')
        tocat.append(outcells)
        # Plot distribution of removal
        plot_sample(outcells, sample, cluster_key, os.path.join(savefig,sample+'_qc.png'))
    allres = pd.concat(tocat)
    return(allres)
```

```python
# Run through the QC procedure across all samples
# This will plot dotplot showing DE genes between outlier and non-outlier groups per cluster/sample
# It is useful to inspect these to make sure that no enrichment for biological relevant genes is being removed
# Generally, this should be highly enriched for MT genes
os.system('mkdir ' + os.path.join(path_aggregation,'humanICM.PreQC.Harmony/nucleiQC'))
warnings.simplefilter(action='ignore', category=FutureWarning)
qc_results = qc_pipeline(adata,
                         'leiden2.0',
                         'sample',
                         ['5', '11', '14', '15', '16', '19', '23',
                          '25', '29', '30', '31', '32', '34', '35',
                          '37', '38', '41', '43', '45', '47'],
                         os.path.join(path_aggregation,'humanICM.PreQC.Harmony/nucleiQC/'))
```
